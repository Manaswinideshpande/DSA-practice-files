1.Given a positive integer n,return the smallest positive integer that is a multiple of both 2 and n

eg 1:
input: n=5
output:10
explanation: the smallest  multiple of both 5 and 2 is 10

Note:*for oddnum:- we do oddnum multiplication with 2 to get the smallest positive integer that is a multiple of both 2 and n


eg 2:
input: n=6
output:6
explanation: the smallest  multiple of both 6 and 2 is 6. note: the number is a multiple of itself

Note: for even we consider evennumber (n) itself to get the smallest positive integer that is a multiple of both 2 and n


```python
n =int(input("enter the value: "))
if n%2 == 0: # even number checking
    print(n)
else:
    print(n*2) # if  odd number multiplied by 2 gives even value
```

***problem3***.Given an array nums containing n distinct numbers in the range[0,n]  return the only number in the range that is missing from array

eg1.
input:nums=[3,0,1]
output:2
explanation:  n=3 since there are 3 numbers , so all numbers are in the range [0,3]. 2 is missing number in the  range since it does not appear in nums.

eg2.
input:nums=[0,1]
output:2
explanation:  n=2 since there are 2 numbers , so all numbers are in the range [0,2]. 2 is missing number in the  range since it does not appear in nums

eg3.
input:nums=[9,6,4,2,3,5,7,0,1]
output:8
explanation:  n=9 since there are 9 numbers , so all numbers are in the range [0,9]. 8 is missing number in the  range since it does not appear in nums

constraints:
n = nums,length
1<=n<=10base4
0<=nums[i]<=n
all the numbers of nums are unique

```python
n = list(map(int, input("Enter numbers separated by space: ").split()))
missing_value = []

for i in range(min(n), max(n)+1):
    if i not in n:
        missing_value.append(i)

print("Missing values:", missing_value)
```
