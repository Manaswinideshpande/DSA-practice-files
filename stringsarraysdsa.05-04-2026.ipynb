Problem 1: Valid Anagram
🧩 Problem Statement

Given two strings s and t, determine if t is an anagram of s.

An anagram is formed by rearranging characters of a string using all original characters exactly once.

🔹 Input
Two strings s and t
🔹 Output
Print "True" if anagram
Else "False"
🔹 Example
Input: listen silent
Output: True
Input: hello world
Output: False
🔹 Constraints
1 ≤ length ≤ 10⁵

```python

from collections import Counter 
a = input("Enter first string: ").strip().lower()
b = input("Enter second string: ").strip().lower()

if Counter(a) == Counter(b):
    print(True)
else:
    print(False)
```

Problem 2: Remove Duplicates from String
🧩 Problem Statement

Given a string, remove duplicate characters while keeping only the first occurrence of each character.

🔹 Input
A string s
🔹 Output
String with duplicates removed
🔹 Example
Input: programming
Output: progamin
🔹 Constraints
1 ≤ length ≤ 10⁵

```python


string = "programming"
new =  ""
for i in string:
    if i not in new:
        new += i
print(new)
    
```

Problem 4: Second Largest Element
🧩 Problem Statement
Given an array of integers, find the second largest distinct element in the array.
🔹 Input
List of integers nums
🔹 Output
Second largest element
🔹 Example
Input: 10 5 20 8
Output: 10
Input: 10 10 5
Output: 5
🔹 Constraints
2 ≤ n ≤ 10⁵

```python
n = list(map(int,input().split()))
max_value1 = float('-inf')
max_value2 = float('-inf')
for i in n:
    if i> max_value1:
        max_value2 = max_value1
        max_value1 = i
    elif i>max_value2 and i!=max_value1:
        max_value2 = i
print(max_value2)
```

Problem 3: Count Vowels in a String
🧩 Problem Statement

Given a string, count the number of vowels (a, e, i, o, u) present in it.

🔹 Input
A string s
🔹 Output
Integer count of vowels
🔹 Example
Input: education
Output: 5
🔹 Constraints
1 ≤ length ≤ 10⁵

```python
# count vowels
v=input("value: ").lower()
n= "aeiou"
count = 0
for i in v:
    if i in n:
        count +=1
print(count)
    
```
