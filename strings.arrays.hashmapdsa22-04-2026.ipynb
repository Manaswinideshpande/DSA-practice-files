```python
Q3: Second Largest Element
```

```python
s = list(map(int,input().split()))
first = float('-inf')
second = float('-inf')
for i in s:
    if i > first:
        second = first
        first  = i
    if i > second and i!=first:
        second = i
print(second)
        
```

```python
Q4: First Non-Repeating Character
```

```python
s = input("")
dic ={}
for i in s:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1

for i in s:
    if dic[i] == 1:
        nonrepeat = i
        break
print(nonrepeat)
    
```
