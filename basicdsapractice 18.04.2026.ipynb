Q2: Find Duplicate Element (again)

👉 Same concept, reinforce

Input: 1 3 4 2 2
Output: 2

```python
s = list(map(int,input().split()))
dic = {}
for i in s:
    if i in dic:
        dic[i] = dic[i] + 1
    else:
        dic[i] = 1
result = None
for i in s:
     if dic[i] > 1:
         result = i
         break
print(result)
```

```python
# method 2
s = list(map(int,input().split()))
set1 = set()
for i in s:
    if i in set1:
        print(i)
        break
    else:
        set1.add(i)
```

Q3: First Unique Character

👉 Same as earlier (important pattern)

Input: aabbcde
Output: c

```python
s = input("")
dic ={}
for i in s:
    if i in dic:
        dic[i] = dic[i] + 1
    else:
        dic[i] = 1
result = None
for i in s:
     if dic[i] == 1:
         result = i
         break
if result:
    print(f"output:{result}")
else:
    print("No unique character found")
    
```

Q4: Count Occurrence of a Target
Input: [1,2,2,3,2], target = 2
Output: 3

```python
s = list(map(int,input().split()))
target = int(input())
count = 0
for i in s:
    if i == target:
        count += 1
print(count)
```

Q5: Check Element Exists
Input: [1,2,3,4], target = 3
Output: True

```python
s = list(map(int,input().split()))
target = int(input())
found = False
for i in s:
    if i == target:
        found = True
        break
print(found)
```
