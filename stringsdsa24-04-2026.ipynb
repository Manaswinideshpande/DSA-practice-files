Q1: Check if Two Arrays are Equal using Hash

Input:
[1,2,3,2]
[2,1,2,3]

Output:
True

hint:
Correct Idea
Need:
Same elements
Same frequency

```python
a = list(map(int,input().split()))
b = list(map(int,input().split()))

freq = {}

for i in a:
    freq[i] = freq.get(i,0) + 1

for j in b:
    if j not in freq:
        print(False)
        break
    freq[j] -= 1
    if freq[j] < 0:
        print(False)
        break
else:
    print(True)
```

Q2: Longest Consecutive Sequence

Input: [100,4,200,1,3,2]

Output: 4
hint:
Sequence: 1,2,3,4

```python
s = list(map(int,input().split()))
nums = set(s)
longest = 0
for i in nums:
    if i - 1 not in nums:
        current = i
        count =1
    while current + 1 in nums:
        current += 1
        count +=1
    longest = max(longest,count)
print(longest)
```
