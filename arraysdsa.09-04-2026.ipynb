problem 1.
given an array of N non- negative integers arr[] representing an elevation map where the width of each bar is 1 
compute how much water it is able to trap after raining

eg1.
input:arr[]= {2,0,2}
output: 2
explanation: the structure is like below we can trap 2 units of water in the middle gap

eg.2
input:arr[]= {3,0,2,0,4}
output: 7
explanation: the structure is like below we can trap 3 units of water between 3 nad 2
" 1 unit on top of bar 2 and 3 units between 2 and 4 "

eg.3
input:arr[]= {0,1,0,2,1,0,1,3,2,1,2,1}
output: 6
explanation: the structure is like below  trap 1 units of water between 1 nad 2
"4 units between second last1 and last2

```python
def trap(arr):
    left = 0
    right = len(arr) - 1
    left_max = 0
    right_max = 0
    water = 0

    while left <= right:
        if arr[left] <= arr[right]:
            if arr[left] >= left_max:
                left_max = arr[left]
            else:
                water += left_max - arr[left]
            left += 1
        else:
            if arr[right] >= right_max:
                right_max = arr[right]
            else:
                water += right_max - arr[right]
            right -= 1

    return water
```

```python
print(trap([3,0,2,0,4]))
```

problem2.

given a boolean 2D 
array, where each row is sorted.find the row with maximum number of 1s.

eg.
    input matrix:  
                  
                  0 1 1 1
                  
                  0 0 1 1
                  
                  1 1 1 1  # this row has max numbers
                  
                  0 0 0 0 


```python
matrix = []

rows = int(input("rows: "))
columns = int(input("columns: "))

# input 2D matrix
for i in range(rows):
    temp = []
    for j in range(columns):
        value = int(input())
        temp.append(value)
    matrix.append(temp)

max_count = 0
row_index = -1

# find row with max 1s
for i in range(rows):
    count = 0   # reset for each row
    for j in range(columns):
        if matrix[i][j] == 1:
            count += 1

    if count > max_count:
        max_count = count
        row_index = i

print("Row with max 1s:", row_index)
print("Number of 1s:", max_count)
    
```
