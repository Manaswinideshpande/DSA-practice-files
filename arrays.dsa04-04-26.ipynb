***problem1***.
given an array of integers nums, return the number  of good pairs
A pair (i,j) is called if nums[i] == nums[j] and i<j

eg1.
input: nums =[1,2,3,1,1,3,3]
output: 4
explanation:
there are 4 good pairs (0,3),(0,4), (3,4),(2,5) 0- indexed .

eg2.

eg3.

```python
nums = list(map(int,input('value:').split()))

count =0
for i in range(0,len(nums)):
    for j in range (0,len(nums)):
        if nums[i] == nums[j]  and i<j:          
            print({i},{j}, 'goodpairs')
            count+=1
 
print(count)
```

```python
#method2
# take list input
nums = list(map(int, input('value: ').split()))

# store frequency in dict
dict1 = {}

for num in nums:
    if num in dict1:
        dict1[num] += 1
    else:
        dict1[num] = 1

print(dict1)

# apply formula n*(n-1)//2 on values and sum
count = 0

for val in dict1.values():
    count += val * (val - 1) // 2 # calculates number of pairs from val items

print(count)
       
```
