***Problem1***.
there are n employees in a company numbered from 0 to n -1.each employee i has worked for hours[i] in the company 
The company requires each employee to work for at least target hours.
you are given a 0- indexed array of non- negative integers hours of length n and  a non- negative integer target 
return the integer denoting the number of employees who worked at least target hours.

example1: 
input: hours =[1,2,3,4],target =2
output : 3
explanation: 
The company wants each employee to work for atleast 2 hrs
Employee 0 worked for 0 hours and did'nt meet the target.
Employee 1 worked for 1 hours and did'nt meet the target.
Employee 2 worked for 2 hours and did'nt meet the target.
Employee 3 worked for 3 hours and did'nt meet the target.
Employee 4 worked for 4 hours and did'nt meet the target.

there are 3 employees who met the target. 

```python
hours = list(map(int,input('value:').split()))
target = int(input('value:'))
count = 0

for i in range(len(hours)):   # each employee
    if hours[i] >= target:   # check their hours
        count += 1
    
print(count)
```

***problem2***. 
Given an array of integers nums and an integer target, return true if there are two numbers
such that they add up to  target.
else return false
you may assume that each input would have exactly one solution, and you may not use the same element twice
eg.1
input: nums=[2,7,11,15] ,target = 9
output: True
Explanation: Because num[0]+nums[1] == 9 , we return True

eg2.
input: nums =[3,2,4] , target = 6
output:True

eg3.
input: nums = [3,3] , target = 7
output: False

constraints:
2<=nums, length <=1e4
-10power9 <= nums[i] <=1e9
-10power9 <= target <=1e9




```python
num = list(map(int,input('value').split()))
target = int(input('value'))
flag = False
for i in range(0,len(num)):
    for j in range(0,len(num)):
        if  i!=j and num[i]+ num[j] == target:
            flag = True
            break
        
    if flag is True:
        break
print(flag)
```
