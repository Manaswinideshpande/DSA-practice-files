you are given strings jewels representing the types of stones that are jewels,and stones representing the stones you have.Each character in stones is a type of stone you have.you want to know how many of the stones you have are also jewels.Letters are case sensitive,so "a" is considered a different type of stone from "A"

eg1.
input: jewels = "aA" , stones = "aAAbbbb"
output : 3


eg2.
input: jewels = "z" , stones = "ZZ"
output : 0


```python
jewel = input('value:' )
stones = input('value:' )
count = 0
for i in jewel:
    for j in stones:
        if i in j:
            count += 1
        
if count ==0:
    print(0)
else:
    print(count)
```

***problem3***.
there is a robot starting at the position(0,0), the origin on a 2d plane Given a sequence of its moves,
judge if this robot ends up at(0,0) after it completes its moves.
you are given a string moves that represents the move sequence of the robot where moves[i] represents its  i th move . valid moves are 'R', 'L' ,'U' ,'D'
return True if the robot returns to the origin after it finishes all of its moves , false otherwise

eg1.
input:moves ='UD'
output =True
explanation:
The robot moves up once, and then down once. All moves  have the same magnitude, so it ended up at the origin where it started. therefore we return true.


eg2.
input:moves ='LL'
output False
explanation:
The robot moves left twice, it ends  up  two moves to the the left of origin. therefore
we return false because it is not at the origin at the end of its move.


```python
moves = input("value: ").upper()

if moves.count('U') == moves.count('D') and moves.count('L') == moves.count('R'):
    print(True)
else:
    print(False)
```

```python
move1 = input("value: ").upper()

count1 = count2 = count3 = count4 = 0

for i in move1:
    if i == 'D':
        count1 += 1
    elif i == 'U':
        count2 += 1
    elif i == 'L':
        count3 += 1
    elif i == 'R':
        count4 += 1

if count1 == count2 and count3 == count4:
    print(True)
else:
    print(False)
```
