Problem 1: Subarray Sum Equals K
Problem Statement

Given an array nums and an integer k, return the total number of subarrays whose sum equals k.

Example

Input:

nums = [1,1,1], k = 2

Output:

2

Explanation:

[1,1] (index 0–1)
[1,1] (index 1–2)
Hint

Use:

prefix sum + hashmap

Key idea:

if (current_sum - k) seen before → subarray exists

```python
nums = list(map(int,input().split()))
k = int(input())
count = 0
for i in range(len(nums)):
    currentsum =0
    for j in range(i,len(nums)):
        currentsum = currentsum+nums[j]
        if currentsum == k:
            count +=1
print(count)   
```

Problem 2: Rotate Array by K Steps
Problem Statement

Given an array nums and an integer k, rotate the array to the right by k steps.

Example

Input:

nums = [1,2,3,4,5,6,7], k = 3

Output:

[5,6,7,1,2,3,4]
Hint
last k elements → move to front

```python
nums = list(map(int,input().split()))
k = int(input())
k = k%len(nums)
result = nums[-k:] + nums[:-k]
print(result)  
```

Problem Statement: 3

Given a string text, check if it is a palindrome after removing spaces.

Example

Input:

text = "nurses run"

Output:

True
Explanation
remove spaces → "nursesrun"
reverse → same → palindrome
Hint
use two pointers OR reverse string

These two will strengthen:

array manipulation + string handling

```python
text = input("").lower()
word = text.replace(" ","")
result = word[::-1]
if word == result:
    print(True)
else:
    print(False)
print(result)
```
