🧩 Problem: Check if a Sentence is a Pangram
📄 Problem Statement

Given a string s representing a sentence, determine whether it is a pangram.

A pangram is a sentence that contains every letter of the English alphabet (a–z) at least once.

The sentence may include:

Uppercase and lowercase letters
Spaces
Digits
Special characters (e.g., #, $, %, @, !)

These non-letter characters should be ignored when checking for a pangram.

🔹 Input
A string s of length 1 ≤ n ≤ 10^5
🔹 Output
Return True if the sentence is a pangram
Return False otherwise

```python
#method1
# take list input
str_ =  "abcdefghijklmnopqrstuvwxyz"
dict1 ={}
for char in str_: # iterates the string values
    if ord(char)>= 97 and ord(char)<=122: # condition to check the presence of a-z chars in a sentence
        # if the  above condition satisfies then store char values in dict
        if char in dict1:
            dict1[char] +=1
        else:
            dict1[char] = 1
if len(dict1) == 26:
    print(True)
else:
    print(False)
```

```python
# method 2
def is_pangram(s):
    s = s.lower()
    letter =set()
    for char in s:
        if 'a'<= char <= 'z':
            letter.add(char)
    return len(letter) == 26
print(is_pangram("The quick brown fox jumps over the lazy dog!"))
print(is_pangram("Python World"))
print(is_pangram("abcdefghijklmnopqrstuvwxyz"))
```

Given two strings a and b consisting of lowercase characters. The task is  to check whether two given strings are an anagram of each others or not.
An anagram of a string is another string that contains the same characters, only the order of characters can be different. 
for example, act and tac are anagram of each other.

eg1.
    input :a= geeksforgeeks ,b=forgeeksgeeks
    output: YES
    explanation: Both the string have same characters with same frequency, So, both are anagrams

eg2.
    input :a= allergy ,b=allergic
    output: No
    explanation: Both the string have different characters So, both are not anagrams

```python
#Same letters + same counts = anagram
from collections import Counter

a = input("Enter first string: ").strip().lower()
b = input("Enter second string: ").strip().lower()

if Counter(a) == Counter(b):
    print("Yes, it is an anagram")
else:
    print("No, it is not an anagram")
```

```python
# method 2
a = input("Enter first string: ").strip().lower()
b = input("Enter second string: ").strip().lower()
count = 0
used =[0]*len(b)
for i in a: 
    for j in range(len(b)):
        if i == b[j] and used[j] ==0:
            count +=1
            used[j] +=1
            break
if count == len(a) and len(a) == len(b):
    print(" Yes, Anagram")
else:
    print("No")
```
