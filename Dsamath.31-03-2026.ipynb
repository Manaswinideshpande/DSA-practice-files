Identify the character that appears most frequently and provide its total count. In cases where multiple characters have the same maximum frequency, any one of them may be returned (or as specified by the constraints).Input FormatA single string $S$ consisting of lowercase English letters, uppercase English letters, digits, or symbols.Output FormatMax Character: The character with the highest frequency.Max Frequency: An integer representing the count of the most frequent character.


Frequency Map: A dictionary or hash map containing all characters in S as keys and their respective counts as values.Example
Input: "abcdaabcdaaa"

Output:Max Character: 'a'

Max Frequency: 6

Frequency Map: {'a': 6, 'b': 2, 'c': 2, 'd': 2}

Constraints 1 <= |S|<= 10 to power5  (The string may contain spaces or special characters.Time Complexity)
Goal: O(n), where n is the length of the string.Space.

***Problem2***:
given an integer array nums,,return  an array answer such that answer[i] is equal to the product of all elements of numbers  except  num[i]
The product of any prefix or suffix of nums is gauranteed  to fit  in a 32 bit integer
you must write an algorithm  that runs in 0(n) time and without using the division operation


example 1:
input: nums = [1,2,3,4]
output: [24,12,8,6]

example 2:
input : nums = [-1,1,0,-3,3]
output: [0,0,9,0,0]

constraints:
2 <= nums, length <=10s
-30 <= nums[i] <=30
the product of any prefix or suffix of nums is gauranteed to fit in a 32 bit- integer



```python
num = [1, 2, 3, 4]
answer = []

for i in range(len(num)):
    product = 1
    for j in range(len(num)):
        if i != j:            # product of all numbers except itself
            product *= num[j]
    answer.append(product)

print(answer)
```

```python
#Method2
num = [1, 2, 3, 4]
answer = []

total_product = 1
for n in num:
    total_product *= n   # 1*2*3*4 = 24

for i in range(len(num)):
    result = total_product // num[i]   # divide by element, not index
    answer.append(result)

print(answer)
```
