1. Maximum Sum Subarray (Kadane’s Algorithm)
Problem Statement

Given an integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

Example

Input:

[-2,1,-3,4,-1,2,1,-5,4]

Output:

6

Explanation:
The subarray with maximum sum is:

[4,-1,2,1]

Sum:

4 + (-1) + 2 + 1 = 6
Approach (Kadane’s Algorithm)

At every element ask:

Should I continue previous subarray?
OR
Should I start a new subarray?
Algorithm
Initialize:
current_sum = nums[0]
max_sum = nums[0]
Traverse remaining elements:
Either start fresh with current element
Or continue previous subarray
Update maximum sum.

```python
arr = list(map(int,input().split()))
current_sum = arr[0]
max_sum = arr[0]
for i in arr[1:]:
    current_sum = max(i,current_sum + i)
    max_sum = max(max_sum ,current_sum) 
print(max_sum)
```

2.Longest Substring Without Repeating Characters
Problem Statement

Given a string text, find the length of the longest substring without repeating characters.

A substring must contain continuous characters.

Example

input = "kitchencatateaabbcc"

Longest substring without repeating characters:

"kitchen"

Length:7
Why?
Progress:

k → ki → kit → kitc → kitch → kitche → kitchen

Length = 7 ✅
Final Output:
max_length = 7

```python
s = input()

seen = set()
left = 0
max_len = 0

for right in range(len(s)):
    while s[right] in seen:
        seen.remove(s[left])  # duplicate removal logic here
        left +=1
      # add current char
    seen.add(s[right])
    max_len = max(max_len,right-left+1)
    # update max length

print(max_len)
```

3. Container With Most Water
Problem Statement

You are given an array height where each element represents the height of a vertical line.

Find two lines that together with the x-axis form a container such that the container stores the maximum amount of water.

Return the maximum water stored.

Example

Input:
[1,8,6,2,5,4,8,3,7]

Output:
49

Formula
Area = width × min(height_left, height_right)

Where:

width = right - left
Explanation

Best pair:

height = 8 and 7

Width:

8 - 1 = 7

Area:

7 × min(8,7)
= 7 × 7
= 49
Approach (Two Pointers)
Place one pointer at start
Place one pointer at end
Calculate area
Move smaller height pointer inward
Repeat until pointers meet

```python
h=[1,8,6,2,5,4,8,3,7]
max_area = 0
left = 0
right = len(h)-1
while left < right:
    width =  right - left
    height = min(h[left],h[right])
    area = width*height
    max_area =max(max_area,area)
    if h[left] < h[right]:
        left +=1
    else:
        right -=1
print(max_area)
        
```
