```python
Q1 (⭐): Best Time to Buy and Sell Stock
Input: [7,1,5,3,6,4]
Output: 5

👉 Buy low, sell high → max profit
```

```python
price = list(map(int,input().split()))
min_val= price[0]
max_profit = 0
for i in price:
    if i < min_val:
        min_val = i
    profit = i - min_val
    max_profit = max(max_profit,profit)
print(max_profit)
print(min_val)
print(profit)
```

```python
Q2: Valid Palindrome (Ignore non-alphanumeric)
Input: "A man, a plan, a canal: Panama"
Output: True
```

```python
s = input("").lower()
clean =""
for i in s:
    if i.isalnum():
        clean  += i
if clean == clean[::-1]:
    print(True)
else:
    print(False)
print(clean)
```

 Problem: Reverse Words in a Sentence

Description:
Given a string, reverse the order of words.
Words are separated by spaces.
 Input:
 hello world from dsa

Output:
dsa from world hello

Constraints:
Words are separated by single space
No leading/trailing spaces

Approach:
Split the string into words
Reverse the list
Join back into string

```python
s = input("").lower()
word =s.split()
reverse = word[::-1]
result =" ".join(reverse)
print(result)
```

 Problem: Remove Duplicates from Array

Description:
Given an array of integers, remove duplicates while preserving order.
 Input:
 1 2 2 3 1 4

Output:
[1, 2, 3, 4]

Constraints:
 Order must be preserved
 Use extra space if needed

Approach:
Use a set to track seen elements
Traverse array and add only unseen elements

```python
w= list(map(int,input().split()))
seen = set()
list1 =[]
for i in w:
    if i not in seen:
        list1.append(i)
        seen.add(i)
print(list1)
```

```python
# Problem: Second Largest Element in an Array

# Description:
# Given an array of integers, find the second largest element.

# Input:
# 3 5 2 9 1

# Output:
# 5

# Constraints:
# - Array has at least 2 elements
# - No sorting preferred (optimal solution)

# Approach:
# - Maintain two variables:
#   first (largest), second (second largest)
# - Update while traversing
```

```python
w= list(map(int,input().split()))
second_val = float("-inf")
first_val = float("-inf")

for i in w:
    if i>first_val:
        second_val = first_val
        first_val = i
    elif i> second_val and i!=first_val:
        second_val = i
print(second_val)
```
