2. The Explanation: The "Outlier Filter"
In Data Science, sometimes you have data that is technically "valid" (it's a number), but it is wrong (an outlier). For example, if you are collecting the ages of high school students and you see "150," someone clearly made a typo.

The Concept: We use a "Double Guard" approach. We check if the data exists (None check) AND if it falls within a logical range (the "Sanity Check").

How the code works:

We set a Minimum and a Maximum limit.

We loop through the data.

We skip None.

We skip anything outside our limits.

```python
raw_ages = ["16", "17", "None", "15", "150", "16"]
clean_ages =[]
for age in raw_ages:
    if age == "None":
        continue
    d_age = int(age)
    if d_age >= 13 and d_age <=20:
        clean_ages.append(d_age)
    else:
        print(f"Removing outliers: {d_age}")
print(f"Cleaned Data: {clean_ages}")
```

3.Task: The "Sensor Guard"
You are monitoring a temperature sensor in a cold-storage room. The temperature should be between -10°C and 5°C.

Your Goal:
Write a program that takes user input (space-separated) and creates a list of "Trusted Temperatures."

Take input using input().split().

Loop through the items.

Clean Step 1: If the item is "None", skip it.

Clean Step 2: Convert the item to an integer.

Clean Step 3: Only add the number to your trusted_temps list if it is between -10 and 5 (inclusive).

If it is outside that range, print "Outlier detected: [number]".

Finally, print the trusted_temps list.

Test Input to try: 2 -15 None 4 20 -8

Why this is useful: This is the exact logic used to "scrub" sensor data before it gets used to train AI models. If you train a model on "20°C" in a freezer, the AI will get confused!

```python
temperature = input("Enter prices (type 'None' for missing): ").split()
trusted_temp =[]
for temp in temperature:
    if temp == "None":
        continue
    try:
        temper = int(temp)
        if temper >= -10 and temper <=5:
            trusted_temp.append(temper)
        else:
            print(f"Outlier value is {temper}")
    except ValueError:
        print(f"Invalid input detected and skipped: '{temp}'")
print(f"cleaned data :{trusted_temp}")
        
```
