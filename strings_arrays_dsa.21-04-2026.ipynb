Q1: Longest Substring Without Repeating Characters ⭐
Input: abcabcbb
Output: 3

👉 Explanation:

"abc" → length = 3 (longest without repeating)
🧠 Hint

👉 Use:

set or hashmap
sliding window (two pointers)

👉 Idea:

Expand window → if duplicate → shrink from left

```python
s = input("")
seen = set()
left = 0
max_len = 0

for right in range(len(s)):
    while s[right] in seen:
        seen.remove(s[left])
        left += 1
    seen.add(s[right])
    max_len = max(max_len, right - left + 1)
print(max_len)
```

```python
Q3: Product of Array Except Self ⭐
Input: [1,2,3,4]
Output: [24,12,8,6]
```

```python
arr = list(map(int,input().split()))
result = []
for i in range(len(arr)):
    product = 1
    for j in range(len(arr)):
        if i!=j:
            product *= arr[j]
    result.append(product)
print(result)     
```

```python
Q4: Move Zeroes (Optimized) ⭐
Input: [0,1,0,3,12]
Output: [1,3,12,0,0]
```

```python
s= list(map(int,input().split()))
new =[]
count =0
for i in s:
    if i != 0:
        new.append(i)
    else:
        count += 1
new.extend([0]*count)
print(new)
```
