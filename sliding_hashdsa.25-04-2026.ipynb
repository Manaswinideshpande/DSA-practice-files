```python
Q1: Longest Substring Without Repeating Characters 
Input: abcabcbb
Output: 3

Goal: fix sliding window fully
```

```python
s = input("")
seen = set()
left = 0
max_len = 0
for right in range(len(s)):
    while s[right] in seen: 
            seen.remove(s[left])
            left +=1
    seen.add(s[right])
    max_len = max(max_len,right - left+1)
print(max_len)
```

```python
Q2: Top K Frequent Elements
Input: [1,1,1,2,2,3], k=2
Output: [1,2]
```

```python
l = list(map(int,input().split()))
k = int(input())

freq ={}
count = 0


for i in l:
    if i in freq:
        freq[i] = freq[i] + 1
    else: 
        freq[i] = 1
result = []
for _ in range(k):
    max_key = None
    max_freq = 0

    for key in freq:
        if freq[key] > max_freq:
            max_freq = freq[key]
            max_key = key

    result.append(max_key)
    del freq[max_key]
print(result)
```
