Problem 1: Check Square Limit
Given an integer n, determine whether the square of n is less than 10⁸.

Return:
True if n² < 100,000,000
False otherwise
🔹 Input
A single integer n
🔹 Output
True or False

```python
n = int(input())

if n * n < 10**8:
    print(True)
else:
    print(False)
```

Problem 2: Reverse an Array
🧩 Problem Statement

Given an array of integers, reverse the array and return the reversed array.

🔹 Input
A list of integers arr
🔹 Output
The reversed array
🔹 Examples
Input: 1 2 3 4
Output: [4, 3, 2, 1]
Input: 5 10 15
Output: [15, 10, 5]

```python
# method1
a =[1,2,3,4] 
result =a[::-1]
print(result)
```

```python
# method2
i = 0
j = len(a) - 1
while i<j:
    a[i],a[j] = a[j],a[i]
    i += 1
    j = -1
print(a)
```

Problem 3: Palindrome String
🧩 Problem Statement

Given a string s, check whether it is a palindrome.

A string is a palindrome if it reads the same forward and backward.

🔹 Input
A string s
🔹 Output
Print "True" if palindrome
Else print "False"

```python
# method 1
n = input("enter the value")
result = n[::-1]
print(result)  
```

```python
#method 2
n = input("enter the value")
rev =""
for i in n:
    rev = i + rev
print(rev)
```

```python
#method 3
n = input("enter the value")
i = 0
j = len(n)-1

while i<j:
    if n[i] != n[j]:
        print("not  palindrome")
        break
    i += 1
    j -=1
else:
    print("palindrome")
```

Problem 4: Move Zeros to End
🧩 Problem Statement

Given an array of integers, move all 0s to the end of the array while maintaining the relative order of non-zero elements.

🔹 Input
A list of integers nums
🔹 Output
Modified list with zeros moved to the end
🔹 Example
Input: 0 1 0 3 12
Output: [1, 3, 12, 0, 0]
🔹 Constraints
1 ≤ n ≤ 10⁵

```python
import builtins
n = list(map(int, builtins.input().split()))
result =[]
for i in n:
    if i != 0:
        result.append(i)

zero = len(n) - len(result)
result.extend([0] * zero)

print(result)
```

Problem 6: Find Largest Element
🧩 Problem Statement

Given an array of integers, find the largest element in the array.

🔹 Input
A list of integers nums
🔹 Output
Print the largest element
🔹 Example
Input: 3 5 2 9 1
Output: 9
🔹 Constraints
1 ≤ n ≤ 10⁵

```python
#nums = list(map(int,input().split()))
nums = [3,5,2,9,1]
max_val = nums[0]

for num in nums:
    if num > max_val:
        max_val = num
print(max_val)
```
